```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
f = open('MID_dist_metres.txt', 'r')

lines = f.readlines()

f.close()
```

```python
# Looking at the text file
for line in lines:
    print(line)
```

```python
dist_from_centre_total = []
dist_from_centre_MeerKAT = []
dist_from_centre_SKA1 = []

for line in lines:
    
    info = line.split()
    dist = (float(info[2]) ** 2 + float(info[3]) ** 2) ** 0.5 / 1000
    dist_from_centre_total.append(dist)
    
    if info[1].startswith('M'):
        dist_from_centre_MeerKAT.append(dist)

    elif info[1].startswith('SKA'):
        dist_from_centre_SKA1.append(dist)        
```

```python
dist_from_centre_total.sort()
dist_from_centre_MeerKAT.sort()
dist_from_centre_SKA1.sort()
```

```python
num_dishes_total = np.arange(len(dist_from_centre_total))
num_dishes_MeerKAT = np.arange(len(dist_from_centre_MeerKAT))
num_dishes_SKA1 = np.arange(len(dist_from_centre_SKA1))
```

```python
plt.plot(dist_from_centre_total, num_dishes_total, color='black', ls='-', lw=3, label='SKA1-Mid total')
plt.plot(dist_from_centre_MeerKAT, num_dishes_MeerKAT, color='red', ls='-', lw=3, label='MeerKAT dishes')
plt.plot(dist_from_centre_SKA1, num_dishes_SKA1, color='royalblue', ls='-', lw=3, label='SKA1 dishes')

plt.xscale('log')

plt.xlabel('Distance from array centre (km)')
plt.ylabel('Number of dishes')
plt.title('Size of SKA1-Mid array with respect to number of dishes')

plt.gcf().set_size_inches(15, 10)
plt.rc('font', size=15)
plt.grid()
plt.legend()
plt.show()
```

```python
f = open('ngvla_dist_metres.txt', 'r')

lines = f.readlines()

f.close()
```

```python
# Looking at the text file
for line in lines:
    print(line)
```

```python
dist_from_centre_total = []
dist_from_centre_main = []
dist_from_centre_lba = []

for i, line in enumerate(lines):
    
    info = line.split()
    dist = (float(info[1]) ** 2 + float(info[2]) ** 2) ** 0.5 / 1000
    dist_from_centre_total.append(dist)
    
    if i < 214:
        dist_from_centre_main.append(dist)
    else:
        dist_from_centre_lba.append(dist)
```

```python
dist_from_centre_total.sort()
dist_from_centre_main.sort()
dist_from_centre_lba.sort()
```

```python
num_dishes_total = np.arange(len(dist_from_centre_total))
num_dishes_main = np.arange(len(dist_from_centre_main))
num_dishes_lba = np.arange(len(dist_from_centre_lba))
```

```python
plt.plot(dist_from_centre_total, num_dishes_total, color='black', ls='-', lw=3, label='Total 18-m dishes')
plt.plot(dist_from_centre_main, num_dishes_main, color='red', ls='-', lw=3, label='Main Interferometric Array dishes')
plt.plot(dist_from_centre_lba, num_dishes_lba, color='royalblue', ls='-', lw=3, label='Long Baseline Array dishes')

plt.xscale('log')

plt.xlabel('Distance from array centre (km)')
plt.ylabel('Number of dishes')
plt.title('Size of ngVLA array with respect to number of dishes')

plt.gcf().set_size_inches(15, 10)
plt.rc('font', size=15)
plt.grid()
plt.legend()
plt.show()
```
